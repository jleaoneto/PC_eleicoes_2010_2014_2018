```
#@title Tabela com relação entre gênero e candidatos
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px 

path = '/content/drive/My Drive/Colab Notebooks/consulta_cand_2010_2014_2018.csv'
df = pd.read_csv(path, sep=';', encoding='latin1')

df_1 = df.groupby(['ANO_ELEICAO','DS_CARGO','DS_GENERO'])['DS_GENERO'].agg( [ ('FREQ','count') ] )
df_1.reset_index(inplace=True)

#color_discrete_sequence=('red','blue')

fig = px.bar(df_1, y='FREQ', facet_col='ANO_ELEICAO', color='DS_GENERO', x='DS_CARGO',
             text='FREQ', barmode = 'stack',
             height=600, width=1700,
             labels={ # replaces default labels by column name
                "FREQ" : "CANDIDATOS", "DS_GENERO" : "GÊNERO", "VR_RECEITA": "Receita",  "DS_CARGO": "", "CD_ELEITO": "Eleito", "ANO_ELEICAO": "Eleição"
            },
             category_orders={ # replaces default order by column name
                "DS_CARGO": ["DEPUTADO ESTADUAL", "DEPUTADO FEDERAL"],
            },
             color_discrete_map={ # replaces default color mapping by value
                                 "FEMININO": "crimson", "MASCULINO": "dodgerblue"
                                 },
             template='presentation'
             )

fig.update_layout(barnorm='percent')

#fig.update_traces(
#    texttemplate="%{y}"
#    )

fig.update_yaxes(
    showgrid=True,
    ticksuffix='%',
    nticks=20
    )

fig.show()
```

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px

path = '/content/drive/My Drive/Colab Notebooks/perfil_eleitorado_2010.csv'
df_2010 = pd.read_csv(path, sep=';', encoding='latin1')

df_2010.query('SG_UF == "RN"', inplace=True)

df_2010 = df_2010.groupby('DS_GENERO', as_index=False)['QT_ELEITORES_PERFIL'].sum()

df_2010['ANO_ELEICAO'] = 'Eleição 2010'

df_2010

path2 = '/content/drive/My Drive/Colab Notebooks/perfil_eleitorado_2014.csv'
df_2014 = pd.read_csv(path2, sep=';', encoding='latin1')

df_2014.query('SG_UF == "RN"', inplace=True)

df_2014 = df_2014.groupby('DS_GENERO', as_index=False)['QT_ELEITORES_PERFIL'].sum()

df_2014['ANO_ELEICAO'] = 'Eleição 2014'

df_2014

path3 = '/content/drive/My Drive/Colab Notebooks/perfil_eleitorado_2018.csv'
df_2018 = pd.read_csv(path3, sep=';', encoding='latin1')

df_2018.query('SG_UF == "RN"', inplace=True)

df_2018 = df_2018.groupby('DS_GENERO', as_index=False)['QT_ELEITORES_PERFIL'].sum()

df_2018['ANO_ELEICAO'] = 'Eleição 2018'

df_2018

df_final = pd.concat( [df_2010, df_2014, df_2018] )

df_final.to_csv('/content/drive/My Drive/Colab Notebooks/perfil_eleitorado_2010_2014_2018.csv', sep=';', encoding='latin1', index=False)


```

```
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import os
from IPython.display import Image

path = '/content/drive/My Drive/Colab Notebooks/perfil_eleitorado_2010_2014_2018.csv'
df = pd.read_csv(path, sep=';', encoding='latin1')

df.query('DS_GENERO == "MASCULINO" or DS_GENERO == "FEMININO"', inplace=True)

fig1 = px.bar(df, y='QT_ELEITORES_PERFIL', color='DS_GENERO', text='QT_ELEITORES_PERFIL', x='ANO_ELEICAO',
             barmode='stack', 
             labels={ # replaces default labels by column name
                "DS_GENERO" : "GÊNERO", "ANO_ELEICAO": "", "QT_ELEITORES_PERFIL" : "Eleitores"
            },             
             color_discrete_map={ # replaces default color mapping by value
                                 "FEMININO": "crimson", "MASCULINO": "dodgerblue"
                                 },
             template='simple_white'
             )
fig1.update_layout(    
    barnorm='percent'
    )
fig1.update_traces(texttemplate="%{y}")
fig1.update_yaxes(
    ticksuffix='%'
    )

# Create subplots: use 'domain' type for Pie subplot
fig2 = make_subplots(rows=1, cols=3, specs=[[{'type':'domain'}, {'type':'domain'}, {'type':'domain'}]],
                     subplot_titles=['Eleição 2010', 'Eleição 2014', 'Eleição 2018']
                     )
fig2.add_trace(go.Pie(labels=df.DS_GENERO, values=df.query('ANO_ELEICAO == "Eleição 2010"')['QT_ELEITORES_PERFIL'], name="Eleição 2010",
              marker_colors=["crimson", "dodgerblue"]), 1, 1)
fig2.add_trace(go.Pie(labels=df.DS_GENERO, values=df.query('ANO_ELEICAO == "Eleição 2014"')['QT_ELEITORES_PERFIL'], name="Eleição 2014",
              marker_colors=["crimson", "dodgerblue"]), 1, 2)
fig2.add_trace(go.Pie(labels=df.DS_GENERO, values=df.query('ANO_ELEICAO == "Eleição 2018"')['QT_ELEITORES_PERFIL'], name="Eleição 2018",
              marker_colors=["crimson", "dodgerblue"]), 1, 3)
#fig2 = px.pie(df, values='QT_ELEITORES_PERFIL', names='DS_GENERO')

fig2.update_layout(template='presentation', 
                   width=900, height=400,                                      
                   )

#if not os.path.exists("images"):
#    os.mkdir("images")

#img_bytes = fig2.to_image(format="png", width=600, height=350, scale=2)
#Image(img_bytes)

fig2.show()

```
